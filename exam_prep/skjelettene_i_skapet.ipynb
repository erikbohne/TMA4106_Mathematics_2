```python
# Definer funksjonen f(x)
def f(x):
    with np.errstate(all='ignore'):  # Håndterer advarsler for divisjon med null
        return np.where(np.abs(x) < 1, np.exp(-1 / (1 - x**2)), 0)

# Sett opp verdiene for x-aksen
x = np.linspace(-1.1, 1.1, 400)

# Beregn y-verdiene ved å bruke funksjonen f(x)
y = f(x)

# Opprett plottet
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='$f(x) = e^{-1/(1-x^2)}$ for $|x|<1$, 0 otherwise')
plt.title('Plot of the Function $f(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-0.1, 1.1)  # Begrenser y-aksen for bedre visuell framstilling
plt.axvline(x=-1, color='r', linestyle='--', label='x = -1')
plt.axvline(x=1, color='g', linestyle='--', label='x = 1')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def u(x, y):
    return x**2 - y**2

def v(x, y):
    return 2*x*y

# Partielle deriverte
def du_dx(x, y):
    return 2*x

def dv_dy(x, y):
    return 2*x

def du_dy(x, y):
    return -2*y

def dv_dx(x, y):
    return 2*y

# Grid for plotting
x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)
X, Y = np.meshgrid(x, y)

U = u(X, Y)
V = v(X, Y)

Du_dx = du_dx(X, Y)
Dv_dy = dv_dy(X, Y)
Du_dy = du_dy(X, Y)
Dv_dx = dv_dx(X, Y)

# Plotting
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.quiver(X, Y, U, V, color='r')
plt.title('Function f(z) = u(x, y) + iv(x, y)')
plt.xlabel('Real Part (u)')
plt.ylabel('Imaginary Part (v)')
plt.axis('equal')

plt.subplot(2, 2, 2)
plt.quiver(X, Y, Du_dx, Dv_dy, color='b')
plt.title('Real derivatives: du/dx and dv/dy')
plt.xlabel('du/dx')
plt.ylabel('dv/dy')
plt.axis('equal')

plt.subplot(2, 2, 3)
plt.quiver(X, Y, Du_dy, Dv_dx, color='g')
plt.title('Imaginary derivatives: du/dy and dv/dx')
plt.xlabel('du/dy')
plt.ylabel('dv/dx')
plt.axis('equal')

plt.subplot(2, 2, 4)
plt.contour(X, Y, Du_dx - Dv_dy, 0, colors='black', linestyles='dashed')
plt.contour(X, Y, Du_dy + Dv_dx, 0, colors='red', linestyles='dotted')
plt.title('Cauchy-Riemann Equations: du/dx=dv/dy, du/dy=-dv/dx')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')

plt.tight_layout()
plt.show()
```
