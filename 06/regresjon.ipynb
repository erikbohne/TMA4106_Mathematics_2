```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x1, x2):
    return x1+2*x2

x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, X2, Z, 50)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title('f(x) = x1+2*x2')
fig.savefig('1.png')
```

```python
def z(x1, x2):
    return 4*x1**2 + 5*x2**2

x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = z(X1, X2)
 
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, X2, Z, 50)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title('f(x) = 4*x1^2 + 5*x2^2')
fig.savefig('2.png')
```

```python
def z(x1, x2):
    return 4*x1**2 + 8*x1 + 5*x2**2 - 10*x2 + 9

x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = z(X1, X2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, X2, Z, 50)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title('f(x) = 4*x1^2 + 8*x1 + 5*x2^2 - 10*x2 + 9')
fig.savefig('3.png')
```

```python
def t(p,V):
    return p*V

p = np.linspace(0, 10, 100)
V = np.linspace(0, 10, 100)
P, V = np.meshgrid(p, V)
T = t(P, V)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(P, V, T, 50)
ax.set_xlabel('P')
ax.set_ylabel('V')
ax.set_title('f(x) = P*V')
fig.savefig('isotherm.png')
```

```python
# Definerer et grid for x og y
x = np.linspace(-1, 3, 400)
y = np.linspace(0, 4, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Punktet (1, 2) og gradienten ved dette punktet
point = np.array([1, 2])
gradient_at_point = np.array([5, 6])

# Vektor vinkelrett på gradienten, som indikerer retningen for å følge ekvidistanselinjen
tangent_vector = np.array([-gradient_at_point[1], gradient_at_point[0]])

# Oppretter plottet
plt.figure(figsize=(8, 6))

# Plotter nivåkurvene for f
contours = plt.contour(X, Y, Z, levels=20, cmap='RdGy')
plt.clabel(contours, inline=True, fontsize=8)

# Markerer punktet (1, 2)
plt.plot(point[0], point[1], 'ro')  # 'ro' for rød sirkel

# Tegner gradientvektoren
plt.quiver(point[0], point[1], gradient_at_point[0], gradient_at_point[1], color='k', scale=1, label='Gradient')

# Tegner den tangentielle vektoren
plt.quiver(point[0], point[1], tangent_vector[0], tangent_vector[1], color='b', scale=1, label='Tangentiell Vektor')

# Setter etiketter og legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Nivåkurver med gradient og tangentiell vektor ved (1, 2)')
plt.legend()
plt.savefig('tangentiell_vektor.png')
```

```python
def f(x, y):
    return x**2 + x*y + y**2 + x + y

# Lager et meshgrid for x og y
x = np.linspace(-2, 3, 400)
y = np.linspace(-1, 4, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Punktet (1, 2) og dets f-verdi
x_point, y_point = 1, 2
z_point = f(x_point, y_point)

# Gradienten ved punktet (1, 2)
grad_f = np.array([5, 6])

# En vektor vinkelrett på gradienten (tangentiell til nivåkurven)
tangent_vector = np.array([-grad_f[1], grad_f[0]])

# Oppretter en 3D-figur
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotter overflaten
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)

# Markerer punktet (1, 2) på overflaten
ax.scatter(x_point, y_point, z_point, color='r', s=50)

# Tegner gradientvektoren
ax.quiver(x_point, y_point, z_point, grad_f[0], grad_f[1], 0, color='k', length=0.5, normalize=True)

# Tegner den tangentielle vektoren
ax.quiver(x_point, y_point, z_point, tangent_vector[0], tangent_vector[1], 0, color='b', length=0.5, normalize=True)

# Tegner den bratteste stigningen nedover fjellsiden
ax.quiver(x_point, y_point, z_point, -grad_f[0], -grad_f[1], 0, color='r', length=0.5, normalize=True)

# Setter etiketter og viser plottet
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
ax.set_title('Fjellsiden med gradient og tangentiell vektor ved (1, 2)')
fig.savefig('fjellsiden.png')
```

```python
# Tegner gradientvektoren som indikerer den bratteste nedstigningen (motsatt retning av gradienten)
plt.figure(figsize=(8, 6))

# Plotter nivåkurvene for f
contours = plt.contour(X, Y, Z, levels=20, cmap='RdGy')
plt.clabel(contours, inline=True, fontsize=8)

# Markerer punktet (1, 2)
plt.plot(point[0], point[1], 'ro')  # 'ro' for rød sirkel

# Tegner en pil motsatt av gradientvektoren for å vise den bratteste nedstigningen
plt.quiver(point[0], point[1], -gradient_at_point[0], -gradient_at_point[1], color='m', scale=10, label='Bratteste Nedstigning (Motsatt Gradient)')

# Setter etiketter og legger til legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Nivåkurver med pil for bratteste nedstigning ved (1, 2)')
plt.legend()
plt.savefig('bratteste_nedstigning.png')

```
